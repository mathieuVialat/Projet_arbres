<h1> Etude des arbres de Grenoble: Partie 2 <h1>

```python
#####################  importer un fichier csv ####################################
import csv
liste_lignes = []
with open("/home/mathieu.vialat@Digital-Grenoble.local/Documents/Projet_arbres/data/ESP_PUBLIC.IDENTITE_ARBRE.csv", newline = '') as f:
    data = csv.reader(f)
    for lignes in data:
        liste_lignes.append(lignes)
print(len(liste_lignes))
print(liste_lignes[5])  
```

```python
############### Identifier tous les genres botaniques de grenoble
############### Combien y en a t-il? 
# genre botanique: colonne 13
```

```python
###################### extraire les genre botanique de tout les arbres sous forme de liste
liste_genre_bota = []
for ind in range(1, len(liste_lignes)):
    liste_genre_bota.append(liste_lignes[ind][13])
print(len(liste_genre_bota))    
print(liste_genre_bota[5])
```

```python
############################ Solution 1
def valeur_uniques_liste(liste):
    liste_unique = []
    for value in liste:
        if (value not in liste_unique and value != ""):
           liste_unique.append(value)
    return liste_unique
```

```python
liste_unique_genre_bota = valeur_uniques_liste(liste_genre_bota)
print(liste_unique_genre_bota)
print("nombre de genre botanique différent:", len(liste_unique_genre_bota))
```

```python
############################ Solution 2
set_genre_bota = set(liste_genre_bota)           
print(len(set_genre_bota))
########################### différence due au catégorie inconnues
set_genre_bota.remove("")
print(len(set_genre_bota))
```

```python
############### Identifier toutes les espèces (espèces + genre) d'arbres de grenoble
liste_genre_espece_bota = []
for ind in range(1, len(liste_lignes)):
    liste_genre_espece_bota.append(liste_lignes[ind][13] + " " + liste_lignes[ind][14])
print(len(liste_genre_espece_bota))    
print(liste_genre_espece_bota[5])
#print(liste_genre_espece_bota)
```

```python
set_genre_espece_bota = set(liste_genre_espece_bota)
print("Le nombre d'espèces d'arbres présentes à Grenoble est: ",len(set_genre_espece_bota))
```

```python
set_genre_espece_bota
```

```python
######## Compter le nombre de représentant de chaque genre botanique en utilisant les structures Dict
def create_dict_from_list(liste):
    Dict = {}
    for value in liste:
        if (value in Dict):
            Dict[value] += 1
        else:
            Dict[value] = 1
    return Dict
```

```python
Dict_genre_bota_nb_arbres = create_dict_from_list(liste_genre_bota)
print(Dict_genre_bota_nb_arbres)
```

```python
################## Genre botanique le plus représenté
def max_dict(Dict):
    maxi = 0
    maxi_nom = ""
    for key, value in Dict.items():
        if(maxi< value):
            maxi= value
            maxi_nom = key
    return maxi, maxi_nom
```

```python
print("Le genre comptant le plus d'arbres à Grenoble est", max_dict(Dict_genre_bota_nb_arbres)[1], "avec", max_dict(Dict_genre_bota_nb_arbres)[0], "arbres.")
```

```python
################## Genre botanique ne comptant qu'un seul représentant
def liste_nb_rep(Dict, nb_rep):
    liste_unique = [key for key, value in Dict.items() if(value==nb_rep)]
    return liste_unique
```

```python
liste_genre_unique = liste_nb_rep(Dict_genre_bota_nb_arbres, 1)
print(liste_genre_unique)
```

```python
################## Genre botanique classé en fonction du nombre d'arbre (décroissant)
sorted_Desc_Dict_genre_bota_nb_arbres = sorted(Dict_genre_bota_nb_arbres.values(), reverse = True)
print(sorted_Desc_Dict_genre_bota_nb_arbres)
```

```python
########################## proportion du genre le plus représenté
print("Le genre d'arbre le plus représenté à Grenoble (", max_dict(Dict_genre_bota_nb_arbres)[1], ") compte pour",max_dict(Dict_genre_bota_nb_arbres)[0]/sum(Dict_genre_bota_nb_arbres.values()), "% du total des arbres")
```

```python
######################## proportion de chacun des genres botaniques par rapport au total
def dict_prop_calc(Dict):
    Dict_prop = {}
    for key, value in Dict.items():
        Dict_prop[key] = Dict[key]/sum(Dict.values())
    return Dict_prop
```

```python
######################## proportion de chacun des genres botaniques par rapport au total
Dict_genre_bota_prop_arbres = dict_prop_calc(Dict_genre_bota_nb_arbres)
print(Dict_genre_bota_prop_arbres)
```

```python
print("Les objectifs de la métro sont remplies car aucun genre ne représente plus de 20% des arbres à lui tout seul!")
```

```python
######## Compter le nombre de représentant de chaque espèce botanique en utilisant les structures Dict
Dict_espece_bota_nb_arbres = create_dict_from_list(liste_genre_espece_bota)
print(Dict_espece_bota_nb_arbres)
```

```python
################## Espece la plus représenté
print("L'espèce comptant le plus d'arbres à Grenoble est", max_dict(Dict_espece_bota_nb_arbres)[1], "avec", max_dict(Dict_espece_bota_nb_arbres)[0], "arbres.")
```

```python
################## Espece ne comptant qu'un seul représentant
liste_espece_unique = liste_nb_rep(Dict_espece_bota_nb_arbres, 1)
print(liste_espece_unique)
```

```python
################## Especes classées en fonction du nombre d'arbre (décroissant)
sorted_Desc_Dict_espece_bota_nb_arbres = sorted(Dict_espece_bota_nb_arbres.values(), reverse = True)
print(sorted_Desc_Dict_espece_bota_nb_arbres)
```

```python
########################## proportion du genre le plus représenté
print("Le genre d'arbre le plus représenté à Grenoble (", max_dict(Dict_espece_bota_nb_arbres)[1], ") compte pour",max_dict(Dict_espece_bota_nb_arbres)[0]/sum(Dict_espece_bota_nb_arbres.values()), "% du total des arbres")
```

```python
######################## proportion de chacun des genres botaniques par rapport au total
Dict_espece_bota_prop_arbres = dict_prop_calc(Dict_espece_bota_nb_arbres)
print(Dict_espece_bota_prop_arbres)
```

<h1>Comment la pratique de plantation a telle évoluée au cours du temps? <h1>
Diversifie t'on plus maintenant qu'avant?



```python
######################### Stockage genre botanique avec annee 
# genre botanique: colonne 13
liste_annee_genre_bota = []
for ind in range(1, len(liste_lignes)):
    liste_annee_genre_bota.append((liste_lignes[ind][19], liste_lignes[ind][13]))
print(len(liste_annee_genre_bota))    
print(liste_annee_genre_bota[5])
#print(liste_annee_genre_bota)
```

```python
######## Compter le nombre de genre botaniques plantés et le nombre d'arbres pour chaque genre en fonction de l'année en utilisant les structures Dict
def fill_dict_withtupList(liste):
    Dict = {}
    for tup in liste:
        if (tup[0] == ""):
            continue
        elif (tup[0] in Dict):
            if (tup[1] in Dict[tup[0]]):
                Dict[tup[0]][tup[1]] += 1
            else:
                Dict[tup[0]][tup[1]] = 1
        else: 
            Dict[tup[0]] = {}
    return Dict
```

```python
Dict_annee_genre_bota = fill_dict_withtupList(liste_annee_genre_bota)
print(Dict_annee_genre_bota)
```

```python
####### trier les arbres par année (croissante)
sorted_Dict_annee_genre_bota = sorted(Dict_annee_genre_bota.keys())
print(sorted_Dict_annee_genre_bota)
```

```python
######################### Stockage espece avec annee 
# genre +espece: colonne 13 + 14
liste_annee_espece = []
for ind in range(1, len(liste_lignes)):
    liste_annee_espece.append((liste_lignes[ind][19], liste_lignes[ind][13] + " " + liste_lignes[ind][14]))
print(len(liste_annee_espece))    
print(liste_annee_espece[5])
#print(liste_annee_espece)
```

```python
######## Compter le nombre d'espèces plantés et le nombre d'arbres pour chaque espèce en fonction de l'année en utilisant les structures Dict
Dict_annee_espece = fill_dict_withtupList(liste_annee_espece)
print(Dict_annee_espece)
```

```python
############# Calcul du nombre de genre et d'espèce par année
liste_nb_genre_annee = []
liste_nb_espece_annee = []
liste_annee = []
## Genre
for annee in Dict_annee_genre_bota:
    liste_nb_genre_annee.append(len(Dict_annee_genre_bota[annee]))
    liste_annee.append(int(annee))
## Espèce  
for annee in Dict_annee_espece:
    liste_nb_espece_annee.append(len(Dict_annee_espece[annee]))
print(liste_nb_genre_annee)
print(liste_nb_espece_annee)
print(liste_annee)
```

```python
#################### plot du nombre de genre différents plantés en fonction de l'année
import matplotlib.pyplot as plt
y_genre = liste_nb_genre_annee
x_genre = liste_annee
plt.scatter(x_genre,y_genre, c='red')
plt.xlabel("Année")
plt.ylabel("Nombre de genres d'arbres différents plantés")
plt.title("Nombre de genres plantés en fonction de l'année")
plt.show()
```

```python
#################### plot du nombre d'espèces différentes plantées en fonction de l'année
y_espece = liste_nb_espece_annee
x_espece = liste_annee
plt.scatter(x_espece,y_espece, c='red')
plt.xlabel("Année")
plt.ylabel("Nombre d'espèces' d'arbres différentes plantés")
plt.title("Nombre d'espèces plantés en fonction de l'année")
plt.show()
```

```python
######################### calcul de la proportion de tout les genres par rapport au total
def dict_dict_prop(Dict_dict):
    Dict_dict_prop = {}
    for key in Dict_dict:
        Dict_dict_prop[key] = {}
        for key2 in Dict_dict[key]:
            Dict_dict_prop[key][key2] = Dict_dict[key][key2]/sum(Dict_dict[key].values())
    return Dict_dict_prop
```

```python
Dict_annee_genre_bota_prop_arbres = dict_dict_prop(Dict_annee_genre_bota)
print(Dict_annee_genre_bota_prop_arbres)
```

```python
######################### calcul de la proportion de toutes les espèces par rapport au total
Dict_annee_espece_prop_arbres = dict_dict_prop(Dict_annee_espece)
print(Dict_annee_espece_prop_arbres)
```

```python
def max_prop_categorie(Dict_dict):
    liste_prop_max = []
    liste_prop_max_nom = []
    for key in Dict_dict:
        maxi = 0
        maxi_nom = ""
        for key2, prop in Dict_dict[key].items():
            if(maxi< prop):
                maxi= prop
                maxi_nom = key2
        liste_prop_max.append(maxi)
        liste_prop_max_nom.append(maxi_nom)
    return liste_prop_max, liste_prop_max_nom
```

```python
liste_prop_genre_max, liste_prop_genre_max_nom = max_prop_categorie(Dict_annee_genre_bota_prop_arbres)
liste_prop_espece_max, liste_prop_espece_max_nom = max_prop_categorie(Dict_annee_espece_prop_arbres)
print(liste_prop_genre_max)
print(liste_prop_genre_max_nom)
print(liste_prop_espece_max)
print(liste_prop_espece_max_nom)
    
```

```python
########################## plot max prop genre en fonction de l'année
y_genre_prop_max = liste_prop_genre_max
x_genre_prop_max = liste_annee
plt.scatter(x_genre_prop_max,y_genre_prop_max, c='black')
plt.xlabel("Année")
plt.ylabel("Proportion du genre le plus planté par rapport au total")
plt.title("Proportion du genre le plus planté en fonction du temps")
plt.plot([1900, 2021], [0.2, 0.2], color = 'red', linestyle = 'dashed')
plt.show()
```

```python
print(" La métro remplie son objectif pour les genres depuis quelques années et la tendance est positive")
```

```python
########################## plot max prop espece en fonction de l'année
y_espece_prop_max = liste_prop_espece_max
x_espece_prop_max = liste_annee
plt.scatter(x_espece_prop_max,y_espece_prop_max, c='blue')
plt.xlabel("Année")
plt.ylabel("Proportion de l'espèce le plus planté par rapport au total")
plt.title("Proportion de l'espèce le plus planté en fonction du temps")
plt.plot([1900, 2021], [0.1, 0.1], color = 'red', linestyle = 'dashed')
plt.show()
```
