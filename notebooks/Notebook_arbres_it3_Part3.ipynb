```python
############################# identifier les essences non désirées
liste_essences_non_desirees = ['Acer saccharinum', 'Albizzia julibrissin', 'Catalpa bignonioïdes', 'Ginko biloba', 'Gleditsia triacanthos', 'Koelreuteria paniculata', 'Laburnum anagyroïdes', 'Paulownia tomentosa', 'Prunus pissardi', 'Sophora japonica', 'Acer platanoides', 'Acer pseudoplatanus', 'Betula costata', 'Betula davurica', 'Betula jacquemontii', 'Betula lenta', 'Betula pendula', 'Betula maximowicziana', 'Betula pubescens', 'Betula utilis', 'Cryptomeria japonica', 'Cupressus cashmeriana', 'Davidia involucrata', 'Fagus sylvatica', 'Juglans nigra', 'Juglans ailantifolia', 'Liriodendron tulipifera', 'Magnolia grandiflora', 'Nerium oleander', 'Picea abies', 'Populus alba', 'Quercus bicolor', 'Quercus macrocarpa', 'Quercus nigra', 'Quercus rubra', 'Quercus alba', 'Salix gracolistyla', 'Salix daphnoides', 'Salix cinerea', 'Salix arenaria', 'Salix alba', 'Salix acutifolia', 'Sorbus aucuparia', 'Ailanthus altissima', 'Robinia pseudoacacia', 'Acer negundo', 'Platanus x acerifolia', 'Acer platanoides', 'Acer pseudoplatanus']
liste_genre_non_desire = ['Aesculus', 'Fraxinus', 'Platanus', 'Betula', 'Castanea', 'Cupressus', 'Olea']

df['ESPECE_COMP'] = df['GENRE_BOTA']+ " " + df['ESPECE']
df = df[df['GENRE'].notna()]
df = df[df['ESPECE'].notna()]
```

```python
df_non_desiree = df[(df['GENRE'].isin(liste_genre_non_desire)) | (df['ESPECE_COMP'].isin(liste_essences_non_desirees))]
df_non_desiree.head()
```

```python
######################### Proportion espèces arbres désirées vs non désirées
print("La proportion d'arbres d'essences 'non désirées' est de ", len(df_non_desiree)/len(df))
```

```python
## Proposer une liste d'espèces à planter qui permet de remplir les objectifs annoncés en termes de diversification
df_desiree = df[(~df['GENRE'].isin(liste_genre_non_desire)) & (~df['ESPECE_COMP'].isin(liste_essences_non_desirees))]
df_desiree.head(10)
```

```python
################## Proportion d'arbes à planter en fonction de l'espèce
serie_prop_espece = df_desiree.groupby('ESPECE_COMP').count()['GENRE']/len(df)
serie_prop_genre = df_desiree.groupby('GENRE_BOTA').count()['GENRE']/len(df)
```

```python
#################### diagramme en barre représentant la proportion des espèces désirées les moins représentées à Grenoble
import matplotlib.pyplot as plt
nb_espece_a_planter = 20
plt.figure(figsize=(18,12))
plt.xlabel('Espèce désirée', fontsize = 25)
plt.ylabel('Proportion', fontsize = 25)
plt.title("Proportion de chaque espèce désirée", fontsize = 25)
serie_prop_espece.sort_values(ascending = False).tail(nb_espece_a_planter).plot(kind = 'bar')
```

```python
#################### diagramme en barre représentant la proportion des espèces désirées les moins représentées à Grenoble
import matplotlib.pyplot as plt
nb_genre_max = 10
plt.figure(figsize=(18,12))
plt.xlabel('Genre désirée', fontsize = 25)
plt.ylabel('Proportion', fontsize = 25)
plt.title("Proportion de chaque genre désirée", fontsize = 25)
serie_prop_genre.sort_values(ascending = False).head(nb_genre_max).plot(kind = 'bar')
```

```python
df_prop_espece = serie_prop_espece.to_frame()
df_prop_espece['ESPECE_COMP'] = df_prop_espece.index
df_prop_espece['GENRE_BOTA'] = df_prop_espece['ESPECE_COMP'].str.split().apply(lambda x: x[0])
df_prop_espece.head(100)
```

```python
################### Liste des espèces désirées les moins plantées à grenoble
#serie_prop_espece.sort_values(ascending = False).tail(nb_espece_a_planter)
serie_prop_genre_sorted_head = serie_prop_genre.sort_values(ascending = False).head(nb_genre_max)
serie_prop_genre_sorted_head
```

```python
df_prop_espece[~df_prop_espece['GENRE_BOTA'].isin(serie_prop_genre_sorted_head.index)]['GENRE'].sort_values(ascending = False).tail(nb_espece_a_planter)
```

```python
########################## Combien d'arbres à planter pour qu'aucune espèce ne dépasse 10% du total comme voulu par la métro
########################## Planter les 20 espèces désirées les moins représentées
############# nb arbres à planter
max_prop = 0.1
nb_total = len(df)
max_espece = df_desiree.groupby('ESPECE_COMP').count().max()['CODE']
nb_total_voulu = max_espece/max_prop
nb_a_planter = nb_total_voulu - nb_total
nb_a_planter_par_espece = nb_a_planter/nb_espece_a_planter
print("Pour diversifier les arbres de la ville de Grenoble en respectant les objectifs de la métro on pourrait planter", int(nb_a_planter_par_espece), " arbres pour les ", nb_espece_a_planter, " espèces de la liste précédente.")
```
