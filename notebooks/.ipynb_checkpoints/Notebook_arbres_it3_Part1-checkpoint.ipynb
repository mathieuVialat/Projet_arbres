```python
############################# stocker conjointement année de plantation et nombre d'arbres plantés
###### déjà fait dans la serie: serie_annee_plantation
```

```python
####################### Quel maire de Grenoble a planté le plus d'arbres
liste_maire = ["Jay", "Rivail", "Viallet", "Cornier", "Mistral", "Martin", "Cocat", "Lafleur", "martin", "Bally", "Perinetti", "Martin", "Michallon", "Dubedout", "Carignon", "Destot", "Piolle"]
annee_maire = [1904, 1908, 1910, 1919, 1932, 1935, 1944, 1945, 1947, 1948, 1949, 1959, 1965, 1983,1995,2014,2021]
print(len(liste_maire), len(annee_maire))
```

```python
Dict_annee_maire = {}
ind_maire = 0
for annee in range(1900, 2022):
    Dict_annee_maire[annee] = liste_maire[ind_maire]
    if (annee >= annee_maire[ind_maire]):
        ind_maire +=1
print(Dict_annee_maire)
```

```python
liste_annee = Dict_annee_maire.keys()
liste_maire_annee = Dict_annee_maire.values()
print(liste_annee)
print(liste_maire_annee)
```

```python
df_maire = pd.DataFrame(list(zip(liste_annee, liste_maire_annee)), columns = ['ANNEEDEPLANTATION', 'MAIRE'])
df_maire.head(50)
```

```python
df_avec_maire = df.merge(df_maire, on = ['ANNEEDEPLANTATION'], how = 'inner')
df_avec_maire.tail(10)
```

```python
df_avec_maire.groupby('MAIRE').count()['CODE'].sort_values(ascending = False)
```

```python
################# Stocker conjointement le genre botanique et l'année de plantation
df_annee_plantation_genre_bota = df[['ANNEEDEPLANTATION', 'GENRE_BOTA']]
df_annee_plantation_genre_bota.tail(10)
```

```python
######################## Pour combien d'arbres le genre botanique est inconnu?
nb_genre_inc = df_annee_plantation_genre_bota['GENRE_BOTA'].isna().sum()
print("On ne connait pas le genre botanique de ", nb_genre_inc, "arbres plantés.")
```

```python
######################## Tracer l'histo du nombre d'arbres plantés par année
df_serie_annee_plantation = serie_annee_plantation.to_frame()
serie_nb_arbres_par_annee = df_serie_annee_plantation.value_counts().sort_index()
#df_serie_nb_arbres_par_annee = serie_nb_arbres_par_annee.to_frame()
serie_nb_arbres_par_annee
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (18,12))
serie_nb_arbres_par_annee.plot(kind = 'bar') #.bar(x = 'ANNEEDEPLANTATION')
```
